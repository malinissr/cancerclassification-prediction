```python
array = data.values
#2dimensional array[row,column]
X = array[:,1:31]
y = array[:,0]
new_array=y.astype('i')

```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
Xs=scaler.fit_transform(X)
```

Training the model


```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=7)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

Principal Component Analysis

```python
from sklearn.decomposition import PCA
n_components=20
pca=PCA(n_components=n_components)
pca.fit(X_train)
pc_train=pd.DataFrame(pca.transform(X_train),columns=['PC'+str(i+1)for i in range(n_components)])
pc_test=pd.DataFrame(pca.transform(X_test),columns=['PC'+str(i+1)for i in range(n_components)])
pca.explained_variance_ratio_
explained_variance=pca.explained_variance_ratio_
explained_variance
```

K-Nearest Neighbor Model

```python
from sklearn.metrics import classification_report,ConfusionMatrixDisplay,confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, y_train)
prediction = knn.predict(X_test)
accuracy = accuracy_score(y_test, prediction)
print("The accuracy of K-Nearest Neighbor model:",accuracy*100)
print(classification_report(y_test,prediction))
```

PCA with KNN model

```python
pca_model=KNeighborsClassifier(n_neighbors=5)
pca_model.fit(pc_train,y_train)
pca_prediction1=pca_model.predict(pc_test)
pca_accuracy1=accuracy_score(y_test,pca_prediction1)
print("The accuracy on PCA KNearestNeighbor model:",pca_accuracy1*100)
```

CONFUSION MATRIX

```python
matrix=confusion_matrix(y_test,prediction)
display=ConfusionMatrixDisplay(confusion_matrix=matrix,display_labels=['Benign','Malignant'])
display.plot(cmap=plt.cm.Blues)
plt.show()
```

PIE CHART

```python
plt.figure(figsize=(7,5))
sns.set_style("whitegrid")
plt.pie(data['diagnosis'].value_counts(),autopct='%1.2f%%', startangle=90)
plt.axis('equal')
plt.title("Benign & Malignant")
plt.show()
```

Support vector Machine Model

```python
from sklearn import svm
support_vector=svm.SVC(kernel='linear',gamma=0.1,C=0.1)
support_vector.fit(X_train,y_train)
prediction2=support_vector.predict(X_test)
accuracy2=accuracy_score(y_test,prediction2)
print("The Accuracy of SVM:",accuracy2*100)
print(classification_report(y_test,prediction2))
```

PCA with Support Vector Machine model

```python
pca_model=svm.SVC(kernel='linear',gamma=0.1,C=0.1)
pca_model.fit(pc_train,y_train)
pca_prediction2=pca_model.predict(pc_test)
pca_accuracy2=accuracy_score(y_test,pca_prediction2)
print("The accuracy on PCA SVM model:",pca_accuracy2*100)
```

Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(max_iter=30000)
lr.fit(X_train,y_train)
prediction1=lr.predict(X_test)
accuracy1=accuracy_score(y_test,prediction1)
print("The accuracy of LogisticRegression model:",accuracy1*100)
print(classification_report(y_test,prediction1))

```

PCA on Logistic Regression Model

```python
pca_model=LogisticRegression(max_iter=30000)
pca_model.fit(pc_train,y_train)
pca_prediction3=pca_model.predict(pc_test)
pca_accuracy3=accuracy_score(y_test,pca_prediction3)
print("The accuracy on PCA LogisticRegression model:",pca_accuracy3*100)
```

Naive Bayes Model

```python
from sklearn.naive_bayes import GaussianNB
NB = GaussianNB()
NB.fit(X_train, y_train)
prediction4=NB.predict(X_test)
accuracy4 = accuracy_score(y_test, prediction4)
print("The accuracy of NaiveBayes model:",accuracy4*100)
print(classification_report(y_test, prediction4))
```

PCA with Naive Bayes Model

```python
pca_model=GaussianNB()
pca_model.fit(pc_train,y_train)
pca_prediction1=pca_model.predict(pc_test)
pca_accuracy1=accuracy_score(y_test,pca_prediction1)
print("The accuracy on PCA GaussianNB model:",pca_accuracy1*100)
```
